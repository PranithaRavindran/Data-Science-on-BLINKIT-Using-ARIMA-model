```
# Cell 1: Import required libraries
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Set random seed for reproducibility
np.random.seed(42)

# Cell 2: Load sample data
def load_data():
    data = {
        'Item_Identifier': ['FD' + str(i) for i in range(1, 8524)],
        'Item_Weight': np.random.uniform(1, 20, 8523),
        'Item_Fat_Content': np.random.choice(['Low Fat', 'Regular'], 8523),
        'Item_Type': np.random.choice(['Fruits and Vegetables', 'Snack Foods', 'Household', 'Frozen Foods', 'Dairy', 'Canned', 'Baking Goods', 'Health and Hygiene'], 8523),
        'Item_MRP': np.random.uniform(30, 250, 8523),
        'Outlet_Identifier': np.random.choice(['OUT01', 'OUT02', 'OUT03', 'OUT04'], 8523),
        'Outlet_Establishment_Year': np.random.choice(range(1985, 2020), 8523),
        'Outlet_Size': np.random.choice(['Small', 'Medium', 'High'], 8523),
        'Outlet_Location_Type': np.random.choice(['Tier 1', 'Tier 2', 'Tier 3'], 8523),
        'Outlet_Type': np.random.choice(['Supermarket Type1', 'Supermarket Type2', 'Supermarket Type3'], 8523),
        'Item_Outlet_Sales': np.random.uniform(100, 10000, 8523)
    }
    return pd.DataFrame(data)

# Cell 3: Calculate KPIs
def calculate_kpis(df):
    kpis = {
        'total_sales': df['Item_Outlet_Sales'].sum(),
        'avg_sales': df['Item_Outlet_Sales'].mean(),
        'total_items': len(df),
        'avg_item_price': df['Item_MRP'].mean(),
        'top_category': df.groupby('Item_Type')['Item_Outlet_Sales'].sum().idxmax()
    }
    return kpis

# Cell 4: Create dashboard with fixed subplot types
def create_dashboard(df, kpis):
    # Create figure with correct subplot specifications
    fig = make_subplots(
        rows=3, cols=2,
        subplot_titles=('Total Sales Trend (1985-2019)', 'Sales by Item Type',
                       'Outlet Size Distribution', 'Sales by Outlet Location',
                       'Sales by Fat Content', 'Sales by Outlet Type'),
        specs=[
            [{"type": "scatter", "colspan": 2}, None],
            [{"type": "bar"}, {"type": "pie"}],
            [{"type": "pie"}, {"type": "bar"}]
        ],
        vertical_spacing=0.1,
        horizontal_spacing=0.05
    )

    # Add KPI cards - fixed annotation format
    kpi_annots = []
    kpi_annots.append(
        dict(
            text=f"<b>Total Sales</b><br>${kpis['total_sales']/1e6:.2f}M",
            xref="paper", yref="paper",
            x=0.02, y=0.98,
            showarrow=False,
            font=dict(size=14),
            align="center",
            bgcolor="rgba(242, 242, 242, 0.8)",
            bordercolor="rgba(0, 0, 0, 0.3)",
            borderwidth=1,
            borderpad=6
        ))
    kpi_annots.append(
        dict(
            text=f"<b>Avg Item Price</b><br>${kpis['avg_item_price']:.2f}",
            xref="paper", yref="paper",
            x=0.25, y=0.98,
            showarrow=False,
            font=dict(size=14),
            align="center",
            bgcolor="rgba(242, 242, 242, 0.8)",
            bordercolor="rgba(0, 0, 0, 0.3)",
            borderwidth=1,
            borderpad=6
        ))
    kpi_annots.append(
        dict(
            text=f"<b>Total Items</b><br>{kpis['total_items']:,}",
            xref="paper", yref="paper",
            x=0.48, y=0.98,
            showarrow=False,
            font=dict(size=14),
            align="center",
            bgcolor="rgba(242, 242, 242, 0.8)",
            bordercolor="rgba(0, 0, 0, 0.3)",
            borderwidth=1,
            borderpad=6
        ))
    kpi_annots.append(
        dict(
            text=f"<b>Top Category</b><br>{kpis['top_category']}",
            xref="paper", yref="paper",
            x=0.71, y=0.98,
            showarrow=False,
            font=dict(size=14),
            align="center",
            bgcolor="rgba(242, 242, 242, 0.8)",
            bordercolor="rgba(0, 0, 0, 0.3)",
            borderwidth=1,
            borderpad=6
        ))

    # Add KPI annotations to the figure
    for annot in kpi_annots:
        fig.add_annotation(annot)

    # Sales Trend (scatter plot)
    yearly_sales = df.groupby('Outlet_Establishment_Year')['Item_Outlet_Sales'].sum().reset_index()
    fig.add_trace(
        go.Scatter(x=yearly_sales['Outlet_Establishment_Year'],
                  y=yearly_sales['Item_Outlet_Sales']/1e6,
                  mode='lines+markers',
                  marker=dict(size=8, color='royalblue'),
                  line=dict(width=3, color='royalblue'),
                  name='Annual Sales'),
        row=1, col=1
    )
    fig.update_xaxes(title_text="Year", row=1, col=1)
    fig.update_yaxes(title_text="Sales (Millions)", row=1, col=1)

    # Sales by Item Type (horizontal bar chart)
    item_sales = df.groupby('Item_Type')['Item_Outlet_Sales'].sum().sort_values(ascending=True)
    fig.add_trace(
        go.Bar(x=item_sales.values/1e6,
               y=item_sales.index,
               orientation='h',
               text=[f"${x/1e6:.2f}M" for x in item_sales.values],
               textposition='auto',
               marker_color='lightseagreen',
               name='Sales by Type'),
        row=2, col=1
    )
    fig.update_xaxes(title_text="Sales (Millions)", row=2, col=1)

    # Outlet Size Distribution (pie chart)
    size_dist = df['Outlet_Size'].value_counts()
    fig.add_trace(
        go.Pie(labels=size_dist.index,
               values=size_dist.values,
               textinfo='percent+label',
               hole=0.4,
               marker=dict(colors=px.colors.qualitative.Pastel),
               name='Outlet Size'),
        row=2, col=2
    )

    # Fat Content Distribution (pie chart)
    fat_dist = df.groupby('Item_Fat_Content')['Item_Outlet_Sales'].sum()
    fig.add_trace(
        go.Pie(labels=fat_dist.index,
               values=fat_dist.values,
               textinfo='percent+label',
               marker=dict(colors=['lightgreen', 'coral']),
               name='Sales by Fat Content'),
        row=3, col=1
    )

    # Sales by Location Distribution (bar chart)
    loc_sales = df.groupby('Outlet_Location_Type')['Item_Outlet_Sales'].sum()
    fig.add_trace(
        go.Bar(x=loc_sales.index,
               y=loc_sales.values/1e6,
               text=[f"${x/1e6:.2f}M" for x in loc_sales.values],
               textposition='auto',
               marker_color='salmon',
               name='Location Sales'),
        row=3, col=2
    )
    fig.update_yaxes(title_text="Sales (Millions)", row=3, col=2)

    # Update layout
    fig.update_layout(
        height=1000,
        showlegend=False,
        title_text="<b>Blinkit Sales Dashboard</b>",
        title_x=0.5,
        title_font=dict(size=24),
        template="plotly_white",
        margin=dict(t=150, b=50, l=50, r=50)
    )

    # Add subtitle
    fig.add_annotation(
        text="<b>Visualizing Product Performance Across Categories and Outlets</b>",
        xref="paper", yref="paper",
        x=0.5, y=0.89,
        showarrow=False,
        font=dict(size=14, color="gray")
    )

    return fig

# Cell 5: Run the dashboard
df = load_data()
kpis = calculate_kpis(df)
fig = create_dashboard(df, kpis)
fig.show()
```
